```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
n = 200
x = np.linspace(-1,1,n+1)
y = np.linspace(-1,1,n+1)

R = 1
```

```python
def f(x,y,R=1.):
    
    z = R**2 - x**2 - y**2
    
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z)
```

```python
f = np.vectorize(f)
```

```python
X,Y = np.meshgrid(x,y)

Z = f(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
def promedio_cuadrado(Z,posicion_x,posicion_y):
    punto_1 = Z[posicion_x][posicion_y]
    punto_2 = Z[posicion_x+1][posicion_y]
    punto_3 = Z[posicion_x][posicion_y+1]
    punto_4 = Z[posicion_x+1][posicion_y+1]
    promedio = (punto_1 + punto_2 + punto_3 + punto_4)/(n**2)
    return promedio
    
```

```python
volumen = 0
for i in range(n):
    for j in range(n):
        volumen += promedio_cuadrado(Z,i,j)
print(volumen)
```

```python
n_3 = 3
raices,pesos = np.polynomial.laguerre.laggauss(n_3)
print(raices)
print(pesos)
```

```python
a = -1
b = 1
f_sustitucion= lambda x : (x**3 * np.exp(x))/((np.exp(x))-1)
t = 0.5*((b-a)*raices + a + b)

integral_numerica = np.sum( pesos*f_sustitucion(t))
integral_real=(np.pi**4)/15
integral_numerica,integral_real
```

```python
error_n = []
for i in range(9):
    raices_n,pesos_n = np.polynomial.laguerre.laggauss(i+2)
    t = 0.5*((b-a)*raices_n + a + b)
    integral_n = np.sum( pesos_n*f_sustitucion(t))
    error = abs(integral_real - integral_n)
    error_n.append(error)
    
print(error_n)
```

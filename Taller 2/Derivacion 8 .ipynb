```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.tan(x))
```

```python
def DerivadaProg(f, x, h):
    return (-3*f(x) + 4*f(x+h) - f(x+2*h))/(2*h)
```

```python
def DerivadaCen(f, x, h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
x, h = np.linspace(0.1, 1.1, 101, retstep = True)
```

```python
fdxProg = DerivadaProg(f, x, h)

plt.plot(x,fdxProg)
```

```python
fdxCen = DerivadaCen(f, x, h)

plt.plot(x,fdxCen)
```

```python
def fdx(x):
    return 1/(np.cos(x)**2*2*(np.tan(x))**(1/2))
```

```python
fdx = fdx(x)

plt.plot(x,fdx)
```

```python
plt.plot(x, np.abs(fdx-fdxCen), label='Derivada central')
plt.plot(x, np.abs(fdx-fdxProg), label='Derivada progresiva')
plt.yscale('log')
plt.title("Error local para derivada progresiva y derivada central")
plt.legend()
```

Se puede observar en la grafica anterior que la derivada central continua siendo mas precisa que la derivada progresiva, incluso al haber calculado una nueva version para esta. No obstante, en terminos generales, se puede decir que ambas derivadas cuentan con un error en el mismo orden de magnitud.

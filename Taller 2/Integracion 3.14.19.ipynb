```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import sympy as sym
```

```python
n_20 = 34
raices,pesos = np.polynomial.legendre.leggauss(n_20)
print(raices)
print(pesos)
```

```python
integral_list = np.zeros(10000)
T = np.linspace(1,20,10000)
a = -1
b = 1
t = 0.5*((b-a)*raices + a + b)
nov = 0.3
diff = T[0]- T[1]
for i in range(1,len(T)):
    dt = diff*300*(1.38)*10**(-23)
    f = lambda x: np.tanh(300*np.sqrt(x**2+dt**2)/(2*T[i]))/(np.sqrt(x**2+dt**2))
    integral_n = (np.sum( pesos*f(t)))/2
    integral_list[i] = integral_n
    if np.abs(integral_n-(1/nov)) < 1e-4:
        print(integral_n)
        temp = T[i]
        print(round(temp,4))

```

```python
plt.vlines(11.8,0,3.3)
plt.plot(T, integral_list)
```

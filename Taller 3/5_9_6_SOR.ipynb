```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
A = np.array([[3,-1,-1],[-1.,3.,1.],[2,1,4]])
b = np.array([1.,3.,7.])
```

```python
def relajacion_sucesiva(A,b,x0,w, itmax=1000, tolerancia= 1e-9):
    
    x = x0.copy()
    
    for it in range(itmax):
        
       
        
        for i in range(A.shape[0]):
            
            sum_ = 0.
            x_j = x[i].copy()  
            for j in range(A.shape[1]):
                if j!=i:
                  sum_ += A[i,j]*x[j]     
            x[i] = (b[i] - sum_)/A[i,i]
            x[i] =(w*x[i])+(1-w)*x_j  
            

            
        residuo = np.linalg.norm( np.dot(A,x) - b )
        
        
        if np.allclose(residuo,tolerancia,atol=tolerancia):
            break
        
    return x,it
```

```python
x0 = np.array([0.,0.,0.])
w = 1.05
a = relajacion_sucesiva(A,b,x0,w)
print(a[0])
```

```python
N = 60
wi = np.linspace(0.5,1,N)
num_iteraciones = np.zeros(N)
min_iterator = 1001
for i in range(N):
  a = relajacion_sucesiva(A,b,x0,wi[i])
  num_iteraciones[i] = int(a[1])
  if int(a[1]) < min_iterator:
    min_iterator = int(a[1])
    best_weight = wi[i]
print(best_weight)
print(min_iterator)
```

```python
plt.plot(wi,num_iteraciones)
```

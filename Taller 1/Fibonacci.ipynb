```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sum(n:int):
    if n == 0:
        x = 0
    elif n == 1:
        x = 1
    else:
        a = n-1
        b= n-2
        x = sum(b) + sum(a)
    return x
```

```python
def n_fibonacci(n):
    x = np.empty(0)
    for p in n:
        p = int(p)
        fibonacci = sum(p)
        x = np.append(x, fibonacci) 
        
    return x
```

```python
x = np.linspace(0,20,21)

y = n_fibonacci(x)
```

```python
plt.plot(x,y)
```

```python
def razon_aurea(n):
    x = np.empty(0)
    for p in n:
        p = int(p) + 1
        q = p + 1
        fibonacci_n_1= sum(q)
        fibonacci_n= sum(p)
        aurea = (fibonacci_n_1/fibonacci_n)
        x = np.append(x, aurea) 
        
    return x
```

```python

plt.plot(x,razon_aurea(x))
plt.axhline(y=((1+np.sqrt(5))/2)) 
```

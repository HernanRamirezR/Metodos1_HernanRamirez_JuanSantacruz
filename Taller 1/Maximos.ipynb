```python
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt"
x = urllib.request.urlopen(url)
filename = 'DatosMaximo.txt'
urllib.request.urlretrieve(url, filename)
datos = np.loadtxt(filename)

posiciones = np.zeros_like((datos))
pendientes = np.diff(datos[:,1])/np.diff(datos[:,0])

i = 0 
while  i < len(pendientes)-1:
    
    if pendientes[i] > 0 and pendientes[i+1] < 0:
        
        posiciones[i][0] = datos[i+1][0]
        posiciones[i][1] = datos[i+1][1]

    i += 1
     
coordenadasX = np.array((0))
coordenadasY = np.array((0))


for i in range(0,len(posiciones)) :
    array = posiciones[i]
    if array.all() != 0:
        coordenadasX = np.append(coordenadasX, posiciones[i][0])
        coordenadasY = np.append(coordenadasY, posiciones[i][1])
        

coordenadasX = np.delete(coordenadasX, 0)
coordenadasY = np.delete(coordenadasY, 0)
    
plt.plot(datos[:,0],datos[:,1])
plt.scatter(coordenadasX, coordenadasY,s=90, color = "orange", marker= "x")
```

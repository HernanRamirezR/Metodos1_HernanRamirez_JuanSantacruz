```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
n = 25
x = np.linspace(-4,4,n)
y = np.linspace(-4,4,n)
X,Y = np.meshgrid(x,y)
```

```python
def potencial_de_flujo(px,py,r=2,V=2):
    potencial = V*px*(1-(r**2/((px)**2+(py)**2)))
    return potencial
```

```python
Vx = np.zeros((n,n))
Vy = np.zeros((n,n))
h=0.001 
for i in range(n):
    for j in range(n):
        r = np.sqrt((x[i])**2+(y[j])**2)
        if r <= 2:
            Vx[i,j] = 0
            Vy[i,j] = 0
        else:
            Vx[i,j] = ((potencial_de_flujo(x[i]+h,y[j]))-(potencial_de_flujo(x[i]-h,y[j])))/(2*h)
            Vy[i,j] = -1*((potencial_de_flujo(x[i],y[j]+h))-(potencial_de_flujo(x[i],y[j]-h)))/(2*h)
            
         

```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
circle = plt.Circle((0,0),2,fill=False,color="blue")
for i in range(n):
    for j in range(n):
        ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j])
ax.add_patch(circle)       
```

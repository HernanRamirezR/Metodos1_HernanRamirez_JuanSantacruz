```python
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
def cumpleaños(N:int):
    
    if N > 365:
        return 1
    else:
        contador = 1
        for i in range(int(N)):
            contador *= (365-i)/(365)
        probabilidad = 1 - contador
        return probabilidad
```

```python
personas = np.linspace(1,80,80)
probabilidad = np.zeros(80)
for i in range(len(probabilidad)):
    probabilidad[i] = cumpleaños(personas[i])
```

```python
plt.plot(personas,probabilidad)
```

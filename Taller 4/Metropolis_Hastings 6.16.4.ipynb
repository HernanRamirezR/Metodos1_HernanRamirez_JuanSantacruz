```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def Cauchy(x):
    return 1 / (np.pi * (1 + x**2))
```

```python
def metropolis(x0, N, sigma= 1):
    sample = np.zeros(N+1)
    sample[0] = x0

    for i in range(0,N):
    
        proposal = sample[i] + np.random.uniform(-sigma, sigma)

        if np.random.rand() < min(1, Cauchy(proposal)/Cauchy(sample[i])):
            sample[i+1] = proposal
        else:
            sample[i+1] = sample[i]
    return sample
```

```python
N = 1000
x0 = 0
samples = metropolis(x0,N)
```

```python
plt.figure()
plt.hist(samples,bins=30,density=True, alpha=0.7, color='b', zorder=2)
x_ = np.linspace(-5, 5, 1000)
plt.plot(x_, Cauchy(x_), 'r', linewidth=2, zorder=3)
plt.legend(['Distribucion Cauchy', 'Distribucion Muestral'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Gauss(x,mu=2,sigma=0.5):
    return np.exp( -0.5*(x-mu)**2/sigma**2  )/np.sqrt(2*np.pi*sigma**2)
```

```python
def metropolis(x0, N, sigma= 0.5):
    sample = np.zeros(N+1)
    sample[0] = x0

    for i in range(0,N):
    
        proposal = sample[i] + np.random.uniform(-sigma, sigma)

        if np.random.rand() < min(1, Gauss(proposal)/Gauss(sample[i])):
            sample[i+1] = proposal
        else:
            sample[i+1] = sample[i]
    return sample
```

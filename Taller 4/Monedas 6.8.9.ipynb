```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def probabilidad_monedas(p1,p2):
        probabilidad = 1*(p1*p2*(1/2)**2)+2*((p1*(1-p2)*(1/2)**2))+ 2*((p2*(1-p1)*(1/2)**2)) + 1*(((1-p1)*(1-p2)*(1/2)**2))
        
        return probabilidad
```

```python
p1 = np.linspace(0.1,0.9,100)
p2 = np.linspace(0.1,0.5,100)
X, Y = np.meshgrid(p1, p2)
Z = probabilidad_monedas(X,Y)
p = probabilidad_monedas(0.9,0.1)
p
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y,Z,cmap="coolwarm")


plt.show()
```

```python
minimo = np.zeros(1)
maximo = np.zeros(1)
posicion_maximo = []
posicion_minimo = []
comparador_maximo = 0
comparador_minimo = 1
for i in range(len(p1)):
    for j in range(len(p2)):
        prob = probabilidad_monedas(p1[i],p2[j])
        if prob > comparador_maximo:
            comparador_maximo = prob
            maximo = prob
            posicion_maximo = [p1[i],p2[j]]
            
            
        if prob < comparador_minimo:
            comparador_minimo = prob
            minimo = prob
            posicion_minimo = [p1[i],p2[j]]
        
```
